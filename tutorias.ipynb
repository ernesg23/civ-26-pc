```python
'''
Se tiene una maquina de sacar juguetes que funciona cuando se ingresa una determinada cantidad de fichas, y se quiere hacer una funcion que imite ese comportamiento.
a. Hacer una funcion que reciba un numero representando el precio de la maquina, e imprima por pantalla "Ingresa x fichas para comenzar" hasta que se hayan ingresado esa cantidad de letras F (que representan una ficha), y luego "¡A jugar!" cuando se hayan ingresado todas las fichas necesarias.
'''
def maquina_juguetes(precio):
    fichas_ingresadas = 0
    while fichas_ingresadas < precio:
        ficha = input(f'Ingresa {precio - fichas_ingresadas} fichas para comenzar: ').upper()
        if ficha == 'F':
            fichas_ingresadas += 1
        else:
            print('NO vale, ingrese una ficha (F)')
    print('¡A jugar!')

maquina_juguetes(5)
```
