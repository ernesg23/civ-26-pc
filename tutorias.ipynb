```python
'''
Se tiene una maquina de sacar juguetes que funciona cuando se ingresa una determinada cantidad de fichas, y se quiere hacer una funcion que imite ese comportamiento.
a. Hacer una funcion que reciba un numero representando el precio de la maquina, e imprima por pantalla "Ingresa x fichas para comenzar" hasta que se hayan ingresado esa cantidad de letras F (que representan una ficha), y luego "¡A jugar!" cuando se hayan ingresado todas las fichas necesarias.
'''
def maquina_juguetes(precio):
    fichas_ingresadas = 0
    while fichas_ingresadas < precio:
        ficha = input(f'Ingresa {precio - fichas_ingresadas} fichas para comenzar: ').upper()
        if ficha == 'F':
            fichas_ingresadas += 1
        else:
            print('NO vale, ingrese una ficha (F)')
    print('¡A jugar!')

maquina_juguetes(5)
```

```python
#3)
productos_comprados = []
def imprimir_productos_comprados(productos):
    total = 0.0
    for i in productos_comprados:
        print(i)
        total += float(i['precio']) * int(i['cantidad'])
    print("Total a pagar: ", total)

def cargar_productos(productos):
    nombre_producto = input("Ingrese el nombre del producto: (* para salir)")
    precio_producto = 0.0
    cantidad_producto = 0
    while nombre_producto != "*":
        str = "Ingrese el precio del producto"
        while True:
            try:
                precio_producto = float(input(str))
            except ValueError:
                str = "Valor incorrecto, ingrese nuevamente el precio"
            else:
                if precio_producto <= 0.0:
                    str = "No puede haber una cantidad nula del producto, ingrese nuevamente la cantidad"
                else:
                    break
        str = "Ingrese la cantidad del producto"
        while True:
            try:
                cantidad_producto = int(input(str))
            except ValueError:
                str = "Valor incorrecto, ingrese nuevamente la cantidad"
            else:
                if cantidad_producto <= 0:
                    str = "No puede haber una cantidad nula del producto, ingrese nuevamente la cantidad"
                else:
                    break

        productos.append(
            {
            "nombre": nombre_producto,
            "precio": precio_producto,
            "cantidad": cantidad_producto
            }
        )

        nombre_producto = input("Ingrese el nombre del producto: (* para salir)")

    return productos

productos_comprados = cargar_productos(productos_comprados)
imprimir_productos_comprados(productos_comprados)
```
